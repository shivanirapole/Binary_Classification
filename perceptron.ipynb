```python
##scatterplot for the dataset
import csv
pointsX10 = [] # points belonging to class 0 and first in pair
pointsX11 = [] # points belonging to class 1 and first in pair
pointsX20 = [] # points belonging to class 0 and second in pair
pointsX21 = [] # points belonging to class 1 and second in pair
num0= 0
num1= 0
with open('dataset_1.csv','r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        if row[3]=="0":
            pointsX10.append(float(row[1]))
            pointsX20.append(float(row[2]))
            num0+=1
        else:
            pointsX11.append(float(row[1]))
            pointsX21.append(float(row[2]))
            num1+=1
            
from matplotlib import pyplot as scttrplt
scttrplt.scatter(pointsX10,pointsX20,color="red") ## red for class 0
scttrplt.scatter(pointsX11,pointsX21,color="blue") ## blue for class 1
scttrplt.show()
```

```python
##perceptron
bias0 = []
bias1 = []
x2 = []
num = num0+num1
## initializing biases and collecting all x2 points in (x1,x2) pair into one list
for i in range (num0):
    bias0.append(0.0)
    x2.append(pointsX20[i])
for i in range (num1):
    bias1.append(0.0)
    x2.append(pointsX21[i])
##initializing weights to 0.8
w = []
c=0.2
for i in range(3):
    w.append(c)
    
def runGradientDescent(w):
    print(w)
    m = -w[1]/w[2]
    c = -w[0]/w[2]
    y = []
    for i in range(num):
        y.append(m*x2[i]+c)
    scttrplt.scatter(pointsX10,pointsX20,color="red")
    scttrplt.scatter(pointsX11,pointsX21,color="blue")
    scttrplt.plot(x2,y, 'black')
    scttrplt.show()
runGradientDescent(w)
```

```python
## checking and performing gradient descent for points in class 0
for i in range(num0):
    val = w[1]*pointsX10[i] + w[2]*pointsX20[i] + w[0] 
    print('val')
    print(val) ## val>0 implies it belongs to class 1
    if(val>0): 
        w[0] = w[0]+ (-1)*0.1
        w[1] = w[1]+ (-1)*pointsX10[i]*0.1
        w[2] = w[2]+ (-1)*pointsX20[i]*0.1
        print(w)
        runGradientDescent(w)
        
        
        
```

```python
## checking and performing gradient descent for points in class 1
for i in range(num1):
    val = w[1]*pointsX20[i] + w[2]+pointsX21[i] + w[0]
    print('val')
    print(val) ## val<=0 implies it belongs to class 0
    if(val<=0):
        w[0] = w[0]+ (1)*0.1
        w[1] = w[1]+ (1)*pointsX10[i]*0.1
        w[2] = w[2]+ (1)*pointsX20[i]*0.1
        print(w)
        runGradientDescent(w)
        
```
